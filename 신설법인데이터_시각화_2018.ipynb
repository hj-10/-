```python
#리스크 임의로 기재된 test 파일 불러오기

state_unemployment ='신설법인_시도_2018.csv'
state_data = pd.read_csv(state_unemployment, encoding = 'utf-8')

state_data.columns = ['code','sido','청년인구','인구대비창업률']
#state_data['지역명'] = state_data.지역명.map(lambda x : str(x).zfill(7))
state_data.head(1)
```

# 다양한 시각화 예시

```python
#OpenStreetMap

m = folium.Map(location=[36, 127], tiles="OpenStreetMap", zoom_start=7)

m.choropleth(
    geo_data=state_geo,
    name='인구대비창업률',
    data=state_data,
    columns=['sido','인구대비창업률'],
    key_on='feature.properties.CTP_KOR_NM',
    fill_color='RdPu',
    fill_opacity=0.7,
    line_opacity=0.3,
    color = 'gray',
    legend_name = '인구대비창업률'
)

folium.LayerControl().add_to(m)

# Save to html
m.save('신설법인_인구대비_2018.html')
webbrowser.open_new("folium_kr.html")

# geo_data : 지도 데이터를 가져와야함
# data : 적용할 판다스 데이터
# columns(a, b) = (지도 데이터와 매핑할 것, 시각화하려는 변수)
# fill_color : 색깔 지정
# key_on : 매핑할 json 변수 지정
```

```python
#"cartoDB positron"

m = folium.Map(location=[36, 127], tiles="cartoDB positron", zoom_start=7)

m.choropleth(
    geo_data=state_geo,
    name='인구대비창업률',
    data=state_data,
    columns=['sido','인구대비창업률'],
    key_on='feature.properties.CTP_KOR_NM',
    fill_color='RdPu',
    fill_opacity=0.7,
    line_opacity=0.3,
    color = 'gray',
    legend_name = '인구대비창업률'
)

folium.LayerControl().add_to(m)

# Save to html
m.save('신설법인_인구대비_2018(gray).html')
webbrowser.open_new("folium_kr.html")

# geo_data : 지도 데이터를 가져와야함
# data : 적용할 판다스 데이터
# columns(a, b) = (지도 데이터와 매핑할 것, 시각화하려는 변수)
# fill_color : 색깔 지정
# key_on : 매핑할 json 변수 지정
```
