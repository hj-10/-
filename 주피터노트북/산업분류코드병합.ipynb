```python
import numpy as np
code_ksic = code[code['산업분류 구분']=='KSIC']
code_ksic
```

```python
code_ksic.info()
```

```python
code_ksic_final = code_ksic[code_ksic['산업분류코드'].map(len) == 3]
code_ksic_final = code_ksic_final.drop(['산업분류 구분','산업분류 구분명'],axis=1)
code_ksic_final.set_index('산업분류코드',inplace=True)
code_ksic_final
```

```python
import pandas as pd

data = pd.read_excel("1.지역별 신설법인 데이터.xlsx")
data
```

```python
data_null=data.dropna(subset=['업종코드_10차'])
data_null
```

```python
def string_cut(x) : 
    return x[:3]

data_null['업종코드_10차_변형'] = data_null['업종코드_10차'].transform(string_cut)
```

```python
data_null.isnull().sum()
```

```python
code_dict = code_ksic_final.to_dict()
type(code_dict)
data_null.replace(code_dict['산업분류명'],inplace=True)
data_null
```

```python
data_null.isnull().sum()
```

```python
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

total_bis = data_null.pivot_table('연령대',index=['업종코드_10차_변형'],aggfunc=['count']).sort_values(('count','연령대'),ascending=False)
total_bis[:20].plot(kind='barh')
```
