```python
import pandas as pd 

data = pd.read_excel("1.지역별 신설법인 데이터.xlsx")
data.head(2)
```

```python
data_backup = data.copy()
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=['시도'],inplace=True)
```

```python
data_sejong = data[data['구군'].isnull()]
```

```python
data.dropna(subset=['구군'],inplace=True)
```

```python
data.isnull().sum()
```

```python
data_gugun = data['구군'].str.split().tolist()
data_gugun_final = []
for gu in data_gugun : 
    data_gugun_final.append(gu[0])
    
data['구군'] = data_gugun_final
data
```

```python
data['설립일자'] = data['설립일자'].apply(lambda x: pd.to_datetime(str(x), format='%Y/%m/%d'))
```

```python
data_2018 = data[data['설립일자'].between('2018-01-01','2018-12-31')]
data_2019 = data[data['설립일자'].between('2019-01-01','2019-12-31')]
data_2020 = data[data['설립일자'].between('2020-01-01','2020-12-31')]
```

```python
data_2018_df = pd.DataFrame(data_2018.groupby(['시도','구군']).size(),columns=['신설법인수'])
data_2018_df
```

```python
data_2019_df = pd.DataFrame(data_2019.groupby(['시도','구군']).size(),columns=['신설법인수'])
data_2019_df
```

```python
data_2020_df = pd.DataFrame(data_2020.groupby(['시도','구군']).size(),columns=['신설법인수'])
data_2020_df
```

```python
data_sejong['설립일자'] = data_sejong['설립일자'].apply(lambda x: pd.to_datetime(str(x), format='%Y/%m/%d'))
```

```python
data_sejong
```

```python
data_sejong_2018 = data_sejong[data_sejong['설립일자'].between('2018-01-01','2018-12-31')]
data_sejong_2019 = data_sejong[data_sejong['설립일자'].between('2019-01-01','2019-12-31')]
data_sejong_2020 = data_sejong[data_sejong['설립일자'].between('2020-01-01','2020-12-31')]
```

```python
data_sejong_2018_df = pd.DataFrame(data_sejong_2018.groupby(['시도']).size(),columns=['신설법인수'])
data_sejong_2018_df.index =['세종시']
data_sejong_2018_df = data_sejong_2018_df.reset_index()
data_sejong_2018_df.columns = ['시도','신설법인수']
data_sejong_2018_df.insert(1,'구군','없음')
data_2018_df = data_2018_df.reset_index().append(data_sejong_2018_df)
data_2018_df
```

```python
data_sejong_2019_df = pd.DataFrame(data_sejong_2019.groupby(['시도']).size(),columns=['신설법인수'])
data_sejong_2019_df.index =['세종시']
data_sejong_2019_df = data_sejong_2019_df.reset_index()
data_sejong_2019_df.columns = ['시도','신설법인수']
data_sejong_2019_df.insert(1,'구군','없음')
data_2019_df = data_2019_df.reset_index().append(data_sejong_2019_df)
data_2019_df
```

```python
data_sejong_2020_df = pd.DataFrame(data_sejong_2020.groupby(['시도']).size(),columns=['신설법인수'])
data_sejong_2020_df.index =['세종시']
data_sejong_2020_df = data_sejong_2020_df.reset_index()
data_sejong_2020_df.columns = ['시도','신설법인수']
data_sejong_2020_df.insert(1,'구군','없음')
data_2020_df = data_2020_df.reset_index().append(data_sejong_2020_df)
data_2020_df
```

```python
#년도별 시군구별 신설법인수
data_2018_df.to_csv('2018시군구별신설법인수.csv',encoding='utf-8-sig')
data_2019_df.to_csv('2019시군구별신설법인수.csv',encoding='utf-8-sig')
data_2020_df.to_csv('2020시군구별신설법인수.csv',encoding='utf-8-sig')
```
