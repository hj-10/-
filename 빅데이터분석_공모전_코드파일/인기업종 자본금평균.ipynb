```python
import pandas as pd 

data = pd.read_excel("1.지역별 신설법인 데이터.xlsx")
data
```

```python
data['설립일자'] = data['설립일자'].apply(lambda x: pd.to_datetime(str(x), format='%Y/%m/%d'))
```

```python
youth_data= data[(data['연령대']=='20대') | (data['연령대']=='30대')]
youth_data
```

```python
youth_data_2018 = youth_data[youth_data['설립일자'].between('2018-01-01','2018-12-31')]
youth_data_2019 = youth_data[youth_data['설립일자'].between('2019-01-01','2019-12-31')]
youth_data_2020 = youth_data[youth_data['설립일자'].between('2020-01-01','2020-12-31')]
```

```python
youth_data_2020.isnull().sum()
```

```python
youth_data_2018 = youth_data_2018.dropna(subset=['업종코드_10차'])
youth_data_2018
youth_data_2019 = youth_data_2019.dropna(subset=['업종코드_10차'])
youth_data_2019
youth_data_2020 = youth_data_2020.dropna(subset=['업종코드_10차'])
youth_data_2020
```

```python
def string_cut(x) : 
    return x[:3]

youth_data_2018['업종코드_10차_변형'] = youth_data_2018['업종코드_10차'].transform(string_cut)
youth_data_2019['업종코드_10차_변형'] = youth_data_2019['업종코드_10차'].transform(string_cut)
youth_data_2020['업종코드_10차_변형'] = youth_data_2020['업종코드_10차'].transform(string_cut)
```

# 산업분류코드 불러와서 inner join

```python
import pandas as pd

code = pd.read_csv("TB_BS_CODE_C.csv",encoding='cp949')
code
```

```python
import numpy as np
code_ksic = code[code['산업분류 구분']=='KSIC']
code_ksic
```

```python
code_ksic_final = code_ksic[code_ksic['산업분류코드'].map(len) == 3]
code_ksic_final = code_ksic_final.drop(['산업분류 구분','산업분류 구분명'],axis=1)
code_ksic_final.set_index('산업분류코드',inplace=True)
code_ksic_final
```

```python
code_dict = code_ksic_final.to_dict()
type(code_dict)

youth_data_2018.replace(code_dict['산업분류명'],inplace=True)

youth_data_2019.replace(code_dict['산업분류명'],inplace=True)

youth_data_2020.replace(code_dict['산업분류명'],inplace=True)
youth_data_2020.head(3)
```

# 2020년 서울 청년 인기업종 평균 자본금

```python
df_seoul = youth_data_2020[youth_data_2020['시도']=='서울']
```

```python
hot_5 = ['의료용 물질 및 의약품 제조업','기타 운송장비 제조업','고무 및 플라스틱제품 제조업','농업', '펄프, 종이 및 종이제품 제조업']
```

```python
hot1 = df_seoul[df_seoul['업종코드_10차_변형']=='의료용 물질 및 의약품 제조업']
hot2 = df_seoul[df_seoul['업종코드_10차_변형']=='기타 운송장비 제조업']
hot3 = df_seoul[df_seoul['업종코드_10차_변형']=='고무 및 플라스틱제품 제조업']
hot4 = df_seoul[df_seoul['업종코드_10차_변형']=='농업']
hot5 = df_seoul[df_seoul['업종코드_10차_변형']=='펄프, 종이 및 종이제품 제조업']
```

```python
hot1_df = top1.groupby('업종코드_10차_변형').aggregate('mean')
hot2_df = top2.groupby('업종코드_10차_변형').aggregate('mean')
hot3_df = top3.groupby('업종코드_10차_변형').aggregate('mean')
hot4_df = top4.groupby('업종코드_10차_변형').aggregate('mean')
hot5_df = top5.groupby('업종코드_10차_변형').aggregate('mean')
```

```python
hot_total = pd.concat([hot1_df,hot2_df,hot3_df,hot4_df,hot5_df],axis=0)
hot_total
```

```python
hot_total['자본금_천원'] = hot_total['자본금']*0.001
```
