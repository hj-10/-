```python
#필요한 패키지 임포트

import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

#지수표현식사용X/
pd.options.display.float_format = '{:.1f}'.format 
```

```python
#파일 불러오기

files = pd.read_excel('2. 지역별 기업현황 데이터(기업이전).xlsx',header=0)
company_move=files.copy()
```

```python
company_move.head()
```

```python
company_move.isnull().sum()
```

```python
#DATETIME 변형

company_move['datetime'] =company_move['기준연월'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m'))
# DataFrame의 index 변형하기
company_move.set_index('datetime', inplace=True) 
# year 컬럼 생성
company_move['year']=company_move.index.year
company_move['month']=company_move.index.month

company_move.reset_index()
```

```python
# 대표자 연령이 20대,30대만 추출

company_move_20dae = company_move.query('대표자연령구간=="20대"')
company_move_30dae = company_move.query('대표자연령구간=="30대"')

# 20대 30대 합치기

company_move_2030dae=pd.concat([company_move_20dae,company_move_30dae])
company_move_2030dae.head()
```

```python
# 결측치 확인 (시도는 결측값이 0)
company_move_2030dae.isnull().sum()
```

```python
# 이전전_구군 누락값 확인해보기

company_move_2030dae[company_move_2030dae['이전전_구군'].isnull()]['이전전_시도'].values
```

```python
# 이전전_구군 결측치 개수 확인
company_move_2030dae[company_move_2030dae['이전전_구군'].isnull()]['이전전_시도'].count()
```

```python
# 이전후_구군 누락값 확인해보기
company_move_2030dae[company_move_2030dae['이전후_구군'].isnull()]['이전후_시도'].values
```

```python
# 이전후_구군 결측치 개수 확인
company_move_2030dae[company_move_2030dae['이전후_구군'].isnull()]['이전후_시도'].count()
```

```python
# 결측치 구군 세종시로 채워주기

company_move_2030dae['이전전_구군']=company_move_2030dae['이전전_구군'].fillna("세종시")
company_move_2030dae['이전후_구군']=company_move_2030dae['이전후_구군'].fillna("세종시")
```

```python
# 결측치 채워졌는지 확인
# 중분류명 결측치 3개 이전전_시도->이전후_시도가 같아서 != 시도 부분에서 자연스럽게 없어짐

company_move_2030dae.isnull().sum()
```

```python
# 이전 시도->이후 시도가 달라진 경우 확인 [시도가 달라진 경우로 시각화 진행]

company_move_2030_diff_sido=company_move_2030dae[company_move_2030dae['이전전_시도']!=company_move_2030dae['이전후_시도']]
company_move_2030_diff_sido
```

```python
# 이전 구군->이후 구군이 달라진 경우 확인

company_move_2030_diff_gugun=company_move_2030dae[company_move_2030dae['이전전_구군']!=company_move_2030dae['이전후_구군']]
company_move_2030_diff_gugun.head()
```

```python
# 이전 구군->이후 구군이 같은 경우 확인(같은 시도,시군군인데 동이 달라진 경우)

company_move_2030dae[company_move_2030dae['이전전_구군']==company_move_2030dae['이전후_구군']].head()
```

```python
# 이전 법정동->이후 법정동이 같은 경우 확인(같은 시도,시군구,동까지 같은데 이전한 경우)

company_move_2030dae[company_move_2030dae['이전전_법정동']==company_move_2030dae['이전후_법정동']].head()
```

```python
# 2018,2019,2020년으로 나누기 [시도가 달라진 경우로 진행 : 

#company_move_2030_diff_sido : 3개년

company_move_2030_diff_sido_18s=company_move_2030_diff_sido.query('year=="2018"')
company_move_2030_diff_sido_18s

company_move_2030_diff_sido_19s=company_move_2030_diff_sido.query('year=="2019"')
company_move_2030_diff_sido_19s

company_move_2030_diff_sido_20s=company_move_2030_diff_sido.query('year=="2020"')
company_move_2030_diff_sido_20s.head()
```

```python
# groupby 하기 위한 컬럼 만들기 

company_move_2030_diff_sido['count']=1 
company_move_2030_diff_sido.head()
```

```python
company_move_2030_diff_sido.isnull().sum()
```

# 기업 이전 시도-> 기업 이후 시도 [year,업종중분류명]


```python
# 기업 이전 시도->기업 이후 시도 year, 업종 별로 그룹바이 하기

company_move_2030_diff_sido_kind=company_move_2030_diff_sido.groupby(['year','이전전_시도','이전후_시도','업종중분류명'])['count'].count()
company_move_2030_diff_sido_kind=pd.DataFrame(company_move_2030_diff_sido_kind)
company_move_2030_diff_sido_kind
```
