```python
#각 연령별로 추출

기업정보_10대=company_info.query('대표자연령구간=="10대"')
기업정보_20대=company_info.query('대표자연령구간=="20대"')
기업정보_30대=company_info.query('대표자연령구간=="30대"')
기업정보_40대=company_info.query('대표자연령구간=="40대"')
기업정보_50대=company_info.query('대표자연령구간=="50대"')
기업정보_60대=company_info.query('대표자연령구간=="60대"')
기업정보_70대=company_info.query('대표자연령구간=="70대"')
기업정보_80대=company_info.query('대표자연령구간=="80대"')
기업정보_90대=company_info.query('대표자연령구간=="90대"')
기업정보_정보없음=company_info.query('대표자연령구간=="정보없음"')
```

```python
#20대30대 추출(매출액 null값 존재)
company_size_youth=pd.concat([기업정보_20대,기업정보_30대],axis=0)
company_size_youth.head()

#20대30대 추출(매출액 null값 존재)데이터에서 기업규모 '미분류','판단제외' 추출
company_size_ex1=company_size_youth.query('기업규모 == "미분류"')
company_size_ex2=company_size_youth.query('기업규모 == "판단제외"')
company_size_excp=pd.concat([company_size_ex1,company_size_ex2])

##20대30대 추출(매출액 null값 존재)데이터에서 기업규모 '미분류','판단제외' 제거
company_size_youth=company_size_youth.drop(company_size_excp,axis=0)
#company_info_youth.query("기업규모=='미분류'")
```

```python
company_size_youth_local=company_size_youth.groupby(['시도','기업규모'])['기업규모'].count()
company_size_youth_local=pd.DataFrame(company_size_youth_local)
company_size_youth_local
```

```python
# 매출액 없는 행 제거

기업정보_10대=기업정보_10대.dropna(subset=['총매출액_천원','평균매출액_천원'])
기업정보_20대=기업정보_20대.dropna(subset=['총매출액_천원','평균매출액_천원'])
기업정보_30대=기업정보_30대.dropna(subset=['총매출액_천원','평균매출액_천원'])
기업정보_40대=기업정보_40대.dropna(subset=['총매출액_천원','평균매출액_천원'])
기업정보_50대=기업정보_50대.dropna(subset=['총매출액_천원','평균매출액_천원'])
기업정보_60대=기업정보_60대.dropna(subset=['총매출액_천원','평균매출액_천원'])
기업정보_70대=기업정보_70대.dropna(subset=['총매출액_천원','평균매출액_천원'])
기업정보_80대=기업정보_80대.dropna(subset=['총매출액_천원','평균매출액_천원'])
기업정보_90대=기업정보_90대.dropna(subset=['총매출액_천원','평균매출액_천원'])
```

```python
# 20~70대 연령대 합치기 

all_age=[기업정보_20대,기업정보_30대,기업정보_40대,기업정보_50대,기업정보_60대,기업정보_70대]
company_age=pd.concat(all_age)
company_age.head()
```

```python
#DATETIME 변형

company_age['datetime'] =company_age['기준연월'].apply(lambda x: pd.to_datetime(str(x), format='%Y/%m/%d'))
# DataFrame의 index 변형하기
company_age.set_index('datetime', inplace=True) 
# year 컬럼 생성
company_age['year']=company_age.index.year
company_age.reset_index()
```
